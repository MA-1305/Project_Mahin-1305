## 1. Ranked actions + reason codes

*The queue: what to do first, and why, in words a human trusts.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

actions = pd.DataFrame({
    "Rank":[1,2,3],
    "Action":[
        "Review Immediately",
        "Monitor Performance",
        "No Immediate Action"
    ],
    "Reason Code":[
        "High Confidence",
        "Medium Confidence",
        "Low Confidence"
    ]
})

print(actions)
```

## Ranked Actions and Reason Codes

The model predictions were converted into a ranked list of recommended actions. Higher confidence predictions receive higher priority for review. These recommendations are intended to support decision-making and should always be verified by a human reviewer.

Reason Codes

• High Confidence → Review Immediately

• Medium Confidence → Monitor Performance

• Low Confidence → No Immediate Action
